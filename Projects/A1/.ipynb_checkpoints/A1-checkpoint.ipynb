## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from scipy.special import gamma
from scipy.stats import beta
import scipy.stats as ss
```

## Maximum Likelihood Estimator & Dataset

```python
uml = np.random.rand()
size = 160
while uml >= 0.4 and uml <= 0.6:
    uml = np.random.rand()
data = np.random.choice([0, 1], size=size, p=[1 - uml, uml])
unique, counts = np.unique(data, return_counts = True)
```

```python
print(f"µml = {counts[1] / 160}")
```

## Prior Distribution

$$𝐵𝑒𝑡𝑎(𝜇| 𝑎, 𝑏) = \frac{𝛤(𝑎 + 𝑏)}{{𝛤(𝑎)}{𝛤(𝑏)}}  𝜇 ^{𝑎−1}  (1 − 𝜇) ^{𝑏−1}$$

$$E[µ] = \frac{a}{a + b} $$

$$var[µ] = \frac {ab}{(a + b)^2(a + b + 1)}$$

__Defining Beta Function__

```python
def beta(x, a, b):
    return (gamma(a + b) / (gamma(a) + gamma(b))) * x**(a - 1) * (1 - x)**(b - 1)
```

__Plotting Prior Distribution__

```python
a , b = 2, 3
x = np.linspace(0, 1, 10000)

fig = plt.figure(figsize = (18, 6))
ax = fig.add_subplot(1, 1, 1)
ax.set_title("Prior Distribution")
ax.plot(x, beta(x, a, b), color = "red", label="p(u|a,b)")
ax.legend();
```

```python
a, b = 2, 3
m, l = 0, 0
x = np.linspace(0, 1, 10000)
fig = plt.figure(figsize = (18, 6))
ax = fig.add_subplot(1, 1, 1)
for i in data:
    if i == 1:
        m += 1
    else:
        l += 1
    y = beta(x, a + m, b + l)
    ax.plot(x, y)
```

```python
a , b = 2, 3
m, l = counts[1], counts[0]
x = np.linspace(0, 1, 10000)
y = ss.beta.pdf(x, a = a + m, b = b + l)
fig = plt.figure(figsize = (20, 6))
ax1 = fig.add_subplot(1, 1, 1)
ax1.set_title("Posterior Distribution")
ax1.plot(x, beta(x, a + m, b + l), color = "blue");
```
